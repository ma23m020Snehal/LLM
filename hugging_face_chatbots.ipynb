Sentiment Analysis alternative way

```python
# defining classifier
classifier = pipeline(task="sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
classifier("he does not like pepsi.")
```

```python
# passing a list to classifier
text_list = ["you should be grateful ", \
             "you have no option ", \
             "i hate this picture"]

classifier(text_list)
```

### Deploying  Chatbot UI

#### Text Sentiment Chatbot

```python
def top3_text_classes(message, history):
    return str(classifier(message)[0][:3]).replace('}, {', '\n').replace('[{', '').replace('}]', '')

demo_sentiment = gr.ChatInterface(top3_text_classes, title="Text Sentiment Chatbot", description="Enter the text, then the chatbot will classify the sentiment.")

demo_sentiment.launch()
```

#### Summarizer Chatbot

```python
def summarizer_bot(message, history):
    return summarizer(message, min_length=5, max_length=140)[0]['summary_text']

demo_summarizer = gr.ChatInterface(summarizer_bot, title="Summarizer Chatbot", description="Enter the text, then chatbot will return the summarized version.")

demo_summarizer.launch()
```

#### Vanilla Chatbot

```python
message_list = []
response_list = []

def vanilla_chatbot(message, history):
    conversation = chatbot(message)

    return conversation[0]['generated_text']

demo_chatbot = gr.ChatInterface(vanilla_chatbot, title="Vanilla Chatbot", description="Enter text to start chatting.")

demo_chatbot.launch()
```
